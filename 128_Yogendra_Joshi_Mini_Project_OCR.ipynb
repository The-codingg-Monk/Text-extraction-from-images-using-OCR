```python
class ProjectPanAadhar():
    
    def __init__(self):
        pass
    def pan_card(self,pan_path):
        import pytesseract
        import re
       
        """
        This function extract texts from PAN card image and provides user info in text.
        """
        if type(pan_path)==str:
            pan_path_new=[]
            pan_path_new.append(pan_path)
            pan_path=pan_path_new  
        pytesseract.pytesseract.tesseract_cmd=r"C:\Program Files\Tesseract-OCR\tesseract.exe"
        dob1=[]
        user=[]
        pan=[]
        for path in pan_path:
            path_list=path.split("\\")
            file_name=path_list[-1]
            file_name=file_name.rstrip(".jpg")
            text=pytesseract.image_to_string(path)
            dob_pattern=r"\b\d{1,2}[-/]\d{1,2}[-/]\d{2,4}\b"
            dob=re.findall(dob_pattern,text)
            dob=dob[0]
            pan_pattern=r"\b[A-Z]{5}[0-9]{4}[A-Z]\b"
            pan_no_list=re.findall(pan_pattern,text)
            try:
                pan_no=pan_no_list[0]
                
                
            except:
                pan_no="Not available"
                
            dob1.append(dob)
            pan.append(pan_no)
            user.append(file_name)
        dict1={"User Name":user,"DOB":dob1,"PAN No.":pan}
      
        dict1={"User Name":user, "PAN No.":pan, "DOB":dob1}
        return dict1
        
    def aadhar_card(self,aadhar_path):
        import pytesseract
        import re
        """
        This function extract texts from aadhar card image and provides user info in text.
        """
        if type(aadhar_path)==str:
            aadhar_path_new=[]
            aadhar_path_new.append(aadhar_path)
            aadhar_path=aadhar_path_new 
        pytesseract.pytesseract.tesseract_cmd=r"C:\Program Files\Tesseract-OCR\tesseract.exe"
        dob_adhar=[]
        adhar=[]
        user_aadhar=[]
        for path in aadhar_path:
            aadhar_dict={}
            path_list=path.split("\\")
            file_name=path_list[-1]
            file_name=file_name.rstrip(".jpg")
            text=pytesseract.image_to_string(path)
            dob_pattern=r"\b\d{1,2}[-/]\d{1,2}[-/]\d{2,4}\b"
            dob_list=re.findall(dob_pattern,text)
            if bool(dob_list):
                dob=dob_list[0]
            else:
                dob_pattern=r"\b[: ]{2,4}[0-9]{4}\b"
                dob_list=re.findall(dob_pattern,text)
                dob_string="".join(dob_list)
                dob_new_list=[]
                for char in dob_string:
                    if char.isdecimal():
                        dob_new_list.append(char)
                dob="".join(dob_new_list)

            aadhar_pattern=r"\b[0-9]{4}[ ][0-9]{4}[ ][0-9]{4}\b"
            aadhar_no_list=re.findall(aadhar_pattern,text)
            try:
                aadhar_no=aadhar_no_list[0]
                
            except:
                aadhar_no="Not available"
                pass
            dob_adhar.append(dob)
            adhar.append(aadhar_no)
            user_aadhar.append(file_name)
            dict2={"User Name":user_aadhar,"DOB":dob_adhar,"Aadhar No.":adhar}
        return dict2



  



import glob
import pandas as pd
pan_path_list=glob.glob(r"C:\Users\kannu\OneDrive\Desktop\DataScience\Projects data\OCR_Project\OCR_data\PAN\*.jpg")
aadhar_path_list=glob.glob(r"C:\Users\kannu\OneDrive\Desktop\DataScience\Projects data\OCR_Project\OCR_data\AADHAR\*.jpg")  
obj=ProjectPanAadhar()
pan_details=obj.pan_card(pan_path_list)
pan_df=pd.DataFrame(pan_details)
pan_df.to_csv("User_PAN_details.csv",index=False)
aadhar_details=obj.aadhar_card(aadhar_path_list)

adhar_df=pd.DataFrame(aadhar_details)
adhar_df.to_csv("User_Aadhar_details.csv",index=False)
print("PAN details are=\n",pan_details)
print("\n\n\n")
print("Aadhar details are=\n",aadhar_details)


   

```
